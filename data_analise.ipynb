```python
#criando uma vari√°el flag para analisar se o fato do balan√ßo estar zerado influencia

BalanceIsZeroData=[]
for a in data["Balance"]:
  if (a == 0.00):
    BalanceIsZeroData.append(1)
  else:
    BalanceIsZeroData.append(0) 


data["BalanceIsZero"] = BalanceIsZeroData

BalanceIsZeroTeste=[]
for a in teste["Balance"]:
  if (a == 0.00):
    BalanceIsZeroTeste.append(1)
  else:
    BalanceIsZeroTeste.append(0) 

teste["BalanceIsZero"] = BalanceIsZeroTeste 
```

Random Forest com as melhores m√©tricas ROC_AUC e Accuracy, por isso vai para o hiperparameter tuning e prediction

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV, KFold
```

```python
#the most important settings are the number of trees in the forest (n_estimators) and the number of features considered for splitting at each leaf node (max_features).
Churn_Model_RFC_Tuned = RandomForestClassifier(random_state=0)
```

```python
espaco_de_parametros = {
 'max_depth': [2, 5, 10, 50, None],
 'max_features': ['auto', 'sqrt'],
 'min_samples_leaf': [1, 2, 4],
 'min_samples_split': [2, 5, 10],
 'n_estimators': [2, 5, 10, 50]}

busca_randomizada = RandomizedSearchCV(Churn_Model_RFC_Tuned,
                              espaco_de_parametros,
                              cv = KFold(n_splits=5),
                                        n_iter=20,
                                      random_state=11)


```

```python
busca_randomizada.fit(X, y)
```

```python
resultados = pd.DataFrame(busca_randomizada.cv_results_)
resultados
```

```python
busca_randomizada.best_estimator_
```

```python
busca_randomizada.best_params_
```

```python
cross_validate(busca_randomizada, X, y, cv=5, scoring=["roc_auc", 'accuracy'])
```
